```python
import re
import emoji
from nltk.tokenize import word_tokenize
from string import punctuation 
from nltk.corpus import stopwords 
import pandas as pd
from langdetect import detect


    
#jli.to_csv('cleanedtweets.csv')

def give_emoji_free_text(text):
    return emoji.get_emoji_regexp().sub(r'', text)


def cleantweet(file):
        data=pd.read_json(file, lines=True,)
        data.to_csv('myjsontweet.csv')

        datac=pd.read_csv("myjsontweet.csv")


        df=datac[['text','created_at']]

        df.dropna(inplace=True)

        li=[]
        _stopwords = set(stopwords.words('english') + list(punctuation) + ['AT_USER','URL'])
        for i,j in zip(df.text,df.created_at):
            a = i.lower() # convert text to lower-case
            a = re.sub('((www\.[^\s]+)|(https?://[^\s]+))', 'URL', a) # remove URLs
            a = re.sub('@[^\s]+123456789', 'AT_USER', a) # remove usernames
            a = re.sub(r'#([^\s]+)!', r'\1', a)
            a=  give_emoji_free_text(a)
            a = word_tokenize(a)
            x=[word for word in a if word not in _stopwords]
            y= ' '.join(word for word in x)
            li.append([y,j])
        #    df.text.replace(to_replace=i,value=y)
        #df
        jli=[]
    
        for i in li:
            try:
                if detect(i[0])=='en':
                    jli.append(i)
                else:
                    continue
            except:
                pass

        jli=pd.DataFrame(jli, columns=['text','timestamp_ms'])
        return jli

path=[]
for i in range(1,15):
    path.append("C:/Users/SHIVAM/jsonfiles/{j}.bz2".format(j=i))

count=1
for i in path:
    x=cleantweet(i)
    x.to_csv('C:/Users/SHIVAM/jsonfiles/tweets{no}.csv'.format(no=count))
    count+=1
```
